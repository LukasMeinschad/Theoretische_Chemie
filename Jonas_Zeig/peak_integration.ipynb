Gaussian Function given by

$g(x) = \frac{1}{\sigma \sqrt{2\pi}}\exp(-\frac{1}{2}\frac{(x-\mu)^2}{\sigma^2})$

+ $\sigma^2$ = variance
+ $\mu$ = expected value

With simple parametric extension

$f(x) = a\exp(-\frac{(x-b)^2}{2c^2})$

```python
import numpy as np
import matplotlib.pyplot as plt


def gaussian_peak(x,a,b,c):
    return a * np.exp(-(x-b)**2/(2*c**2))

x_vals = np.linspace(-10,10,1000)

a = 1
b = 0
c = 1

y = gaussian_peak(x_vals,a,b,c,)

noise = np.random.normal(0,0.05,y.shape)
y_noise = y + noise

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y, "b--",label ="Normal Gaussian Peak")
plt.plot(x_vals,y_noise, "r--",label="Gaussian Peak with Noise")
plt.legend()
plt.show()
```

Calculate the Numerical Gradient (Talk with Jonas which Method to Choose here)

```python
y_prime = np.gradient(y,x_vals)
y_prime_noise = np.gradient(y_noise,x_vals)

plt.plot(x_vals,y_prime, "b--", label="1. Derivative, without noise")

plt.plot(x_vals,y_prime_noise, "r--",label = "1. Derivative, with noise")
plt.legend()
plt.show()
```

### Small Test on more assymetric gaussian peaks

```python
def asymmetric_gaussian_peak(x, a, b, c, d):
    return a * np.exp(-(x-b)**2 / (2*c**2)) * (1 + d * (x-b))

x_vals = np.linspace(-10, 10, 1000)

a = 1
b = 0
c = 1
d = 0.6

y = asymmetric_gaussian_peak(x_vals, a, b, c, d)

noise = np.random.normal(0, 0.05, y.shape)
y_noise = y + noise

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y, "b--", label="Asymmetric Gaussian Peak")
#plt.plot(x_vals, y_noise, "r--", label="Asymmetric Gaussian Peak with Noise")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Asymmetric Gaussian Peak')
plt.show()
```
